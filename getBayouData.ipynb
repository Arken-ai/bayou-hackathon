```python
def getFullCustomerInfo(id):
    customer = requests.get(f"https://{bayou_domain}/api/v2/customers/{id}", auth=(BAYOU_API_TOKEN, '')).json()
    bills = requests.get(f"https://{bayou_domain}/api/v2/customers/{id}/bills", auth=(BAYOU_API_TOKEN, '')).json()
    try:
        billsDF=pd.DataFrame(bills)
        totalUsage = billsDF["electricity_consumption"].sum()
        avgUsage = billsDF["electricity_consumption"].mean()
        totalCost = billsDF["total_amount"].sum()
        avgCost = billsDF["total_amount"].mean()
        customer["bills"]=bills
        return {
            "id": customer["id"],
            "customer":customer,
            "totalUsage":totalUsage,
            "avgUsagePerMonth":avgUsage,
            "totalCost":totalCost,
            "avgCostPerMonth":avgCost
        }
    except Exception as err:
        print(err)
        return {
            "id": customer["id"],
            "customer":customer,
        }
```

```python
customerInfo = getFullCustomerInfo(2284)
customerBills = pd.DataFrame(customerInfo["customer"]["bills"])
sns.boxplot(x=customerBills["electricity_consumption"] );
```
